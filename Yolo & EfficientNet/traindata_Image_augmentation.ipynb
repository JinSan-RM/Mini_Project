```python
import os
from PIL import Image
import random
import glob

# 검은색을 투명하게 만들고, 동시에 회색계열로 조정하여 augment하는 함수
def augment_image(image, transparency_factor_range=(0.4, 1.0), brightness_factor_range=(0.4, 1.0)):
    transparency_factor = random.uniform(*transparency_factor_range)
    brightness_factor = random.uniform(*brightness_factor_range)

    image = image.convert("RGBA")  # RGBA 모드로 변환
    data = image.getdata()

    new_data = []
    for item in data:
        # 검은색 픽셀을 투명하게 만들기
        if item[:3] == (0, 0, 0):
            new_pixel = (0, 0, 0, int(255 * transparency_factor))
        else:
            # 검은색 픽셀의 RGB 값을 수정하여 회색계열로 변형
            new_pixel = (
                int(item[0] * brightness_factor),
                int(item[1] * brightness_factor),
                int(item[2] * brightness_factor),
                item[3]
            )
        new_data.append(new_pixel)

    image.putdata(new_data)
    return image

# 이미지 파일들이 있는 폴더 경로
folder_path = '/tf/notebook/icon/insta/'

# 부풀릴 이미지 개수
augmented_count = 5

# 폴더 내의 모든 이미지 파일들 읽어오기
image_files = glob.glob(os.path.join(folder_path, '*.png')) + glob.glob(os.path.join(folder_path, '*.jpg'))

# 이미지 부풀리기
for file_path in image_files:
    # 이미지 열기
    image = Image.open(file_path)

    for i in range(augmented_count):
        # 이미지를 투명하게 만들고, 동시에 회색계열로 조정하여 한 번에 augment
        augmented_image = augment_image(
            image,
            transparency_factor_range=(0.3, 1.0),  # 투명도를 0부터 1까지 랜덤하게 설정
            brightness_factor_range=(0.5, 1.0)    # 회색계열 정도를 0.5부터 1까지 랜덤하게 설정
        )

        # 부풀린 이미지 파일 저장
        file_name = os.path.basename(file_path)
        new_file_name = f"augmented_{i+1}_{file_name}"
        new_image_path = os.path.join(folder_path, new_file_name)
        augmented_image.save(new_image_path)

        # 이미지 부풀리기 결과 출력
        print(f"Created augmented image: {new_file_name}")

    # 이미지 닫기
    image.close()
```
