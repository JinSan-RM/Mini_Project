```python
import os

folder1 = 'C:/Users/leeyo/Project/test/images_val'
folder2 = 'C:/Users/leeyo/Project/test/label_text'

files1 = os.listdir(folder1)
files2 = os.listdir(folder2)

common_files = []
different_files = []

for file1 in files1:
    name1, ext1 = os.path.splitext(file1)
    found = False
    for file2 in files2:
        name2, ext2 = os.path.splitext(file2)
        if name1 == name2:
            common_files.append(file1)
            found = True
            break
    if not found:
        different_files.append(file1)

for file2 in files2:
    name2, ext2 = os.path.splitext(file2)
    found = False
    for file1 in files1:
        name1, ext1 = os.path.splitext(file1)
        if name1 == name2:
            found = True
            break
    if not found:
        different_files.append(file2)

print("Common files:")
for file in common_files:
    print(file)

print("Different files:")
for file in different_files:
    print(file)
```
